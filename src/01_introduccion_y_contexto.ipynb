```python
# notebooks/01_introduccion_y_contexto.ipynb

import gymnasium
import numpy as np
import matplotlib.pyplot as plt
import os
import sys

# AÃ±ade la carpeta src al path para poder importar utils.py
src_path = os.path.abspath(os.path.join("..", "src"))
if src_path not in sys.path:
    sys.path.append(src_path)

from utils import run_and_render_episode, random_policy
```

```python
# Crea el entorno con render_mode que devuelve arrays RGB
# Entorno de Blackjack sin las reglas naturales, escenario mÃ¡s parecido a un casino y sin reglas de Sutton & Barto
env = gymnasium.make("Blackjack-v1", render_mode="rgb_array", natural=False, sab=False)

# Ejecuta y muestra 3 episodios
for ep in range(3):
    frames, total_reward = run_and_render_episode(env, random_policy)
    n = len(frames)
    print(f"=== Episodio {ep+1}: {n} pasos ===")

    # Mostrar frames
    for i, frame in enumerate(frames):
        plt.figure(figsize=(4, 3))
        plt.imshow(frame)
        plt.axis("off")
        plt.title(f"Episodio {ep+1}, Paso {i+1}")
        plt.show()

    # Resultado final del episodio
    if np.isclose(total_reward, 1.5):
        result = "Â¡Ganaste con Blackjack Natural!"
    elif np.isclose(total_reward, 1):
        result = "Â¡Ganaste!"
    elif np.isclose(total_reward, -1):
        result = "Perdiste"
    elif np.isclose(total_reward, 0):
        result = "Empate"
    else:
        result = f"Recompensa inesperada: {total_reward}"

    print(f"Resultado final del Episodio {ep+1}: {result}")
```
